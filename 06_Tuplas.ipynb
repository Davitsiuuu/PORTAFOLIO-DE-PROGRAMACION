##TUPLAS

```python
my_set = set()
other_set = {}
print(type(my_set))
print(type(other_set))
```

```python
other_set = { 1, 2.5, 3, 'Abdiel', 'Said', 'Hernández', 'Martínez'}
print(type(other_set))
```

##LONGITUD

```python
my_set = { 'banana', 'mango', 'naranja', 'limon' }
print(len(my_set))
print(my_set)
```

##AGREGAR ITEMS

```python
print(my_set)
my_set.add('manzana')
print(my_set))
```

##AGREGAR MULTIPLES ITEMS

```python
print(my_set)
vegetales = {'tomate', 'papa', 'zanahoria'}
print(vegetales)
my_set.update(vegetales)
print(my_set)
```

##REVISAR POSICIONES

```python
frutas = { 'banana', 'mango', 'naranja', 'limon'}
print (frutas)
print(frutas[0])
```

```python
print('¿Existe mango?', 'mango' in frutas)
```

##REMOVER ELEMENTOS

```python
print(frutas)
frutas.remove(frutas[0])
```

```python
frutas.remove('mango')
print(frutas)
```

##UNION

```python
frutas = {'banana', 'limon', 'mango', 'naranja'}
vegetales = {'tomate', 'papa', 'zanahoria'}
print(frutas.union(vegetales))
```
