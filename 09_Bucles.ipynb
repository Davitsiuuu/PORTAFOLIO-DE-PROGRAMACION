##*FOR*
Su función es similar a la utilizada

```python
person = {
    'name' : 'Abdiel',
    'last_name' : 'Hernández',
    'Age' : 19,
    'skills' : ['Javascript', 'react', 'Python', 'switf'],
    'address' : {
        'street' : 'Fco. I. Madero',
        'teléfono' : 7821004745
    }
}

for key in person
  print(key)

for key, value in person.items():
  print(' \'key\' :',key,' \'value:\' ', value)
```
